**Lottery game using random number generation**

```python
import random

comp=random.randint(1,9)

count1=0 #Computer counter
count2=0 #User Counter
for x in range(3):
  user=int(input(" Enter the number between 0 & 9 "))
  print("Computer generated=",comp)
  print("User input=",user)
  if comp>user:
    print("Computer Won")
    count1+=1
  elif comp==user:
    print("TIE")
    count1+=1
    count2+=1
  else:
    print("You Won")
    count2+=1

print("Computer won",count1,"times")
print("You won",count2,"times")


```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=4+2*np.sin(x)

fig, a = plt.subplots()
a.plot(x,y,linewidth=3.0,color='red')

a.set(xlim=(0,8), xticks=np.arange(1,8),
      ylim=(0,8), yticks=np.arange(1,8))
plt.title("sine wave",loc='left')
plt.grid()
plt.show()
```

**Scatter plot**
*Ex-Temperature vs resistance*

```python
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10, 20, 25, 30, 40, 50, 60, 70, 80, 90, 100, 110, 120])
Res=np.array([18476, 12185, 10000, 8260,5740, 4080, 2960, 2188, 1645,1257,973,765,608])

plt.scatter (Temp, Res,color='red')
plt.grid()
plt.title("Temperature vs resistance")
plt.xlabel("Temperature")
plt.ylabel("Resistance")
plt.show()
```

**using plot instead of scatter in the above program!**

```python
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10, 20, 25, 30, 40, 50, 60, 70, 80, 90, 100, 110, 120])
Res=np.array([18476, 12185, 10000, 8260,5740, 4080, 2960, 2188, 1645,1257,973,765,608])

plt.plot (Res,marker='*')
plt.grid()
plt.title("Temperature vs resistance")
plt.xlabel("Temperature")
plt.ylabel("Resistance")
plt.show()
```

```python
import random

comp=random.randint(1,9)

count1=0 #Computer counter
count2=0 #User Counter
count3=0 #TIE
for x in range(3):
  user=int(input(" Enter the number between 0 & 9 "))
  print("Computer generated=",comp)
  print("User input=",user)
  if comp>user:
    print("Computer Won")
    count1+=1
  elif comp==user:
    print("TIE")
    count3+=1
  else:
    print("You Won")
    count2+=1

print("Computer won",count1,"times")
print("You won",count2,"times")

# Calculate success rates
total_games = count1 + count2 + count3
success_rate_count1 = count1 / total_games
success_rate_count2 = count2 / total_games
success_rate_count3 = count3 / total_games

# Create a bar plot
labels = ['Comp', 'User', 'TIE']
success_rates = [success_rate_count1, success_rate_count2, success_rate_count3]
colors = ['blue', 'green', 'orange']

plt.bar(labels, success_rates, color=colors)
plt.xlabel('Counts')
plt.ylabel('Success Rate')
plt.title('Success Rate of Game Counts')
plt.ylim(0, 1)  # Set y-axis limits to ensure proper visualization
plt.show()



```
